```python
import requests
import pandas as pd
from pathlib import Path

def get_btc_5min(save_path: str | Path = "btc_5min.parquet") -> pd.DataFrame:
    """
    Return the latest 5-minute BTC/USD candlesticks from EODHD
    and save them to disk.
    """
    url = (
        "https://eodhd.com/api/intraday/"
        f"BTC-USD.CC?api_token={EODHD_token}&interval=5m&fmt=json"
    )

    r = requests.get(url, timeout=10)
    r.raise_for_status()

    candles = r.json()
    if not candles:
        raise ValueError("No candlestick data returned.")

    df = pd.DataFrame(candles)
    df["datetime"] = pd.to_datetime(df["datetime"], utc=True)
    df = df.set_index("datetime").sort_index()

    num_cols = ["open", "high", "low", "close", "volume"]
    df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")

    # Save to disk
    save_path = Path(save_path)
    if save_path.suffix == ".csv":
        df.to_csv(save_path)
    else:  # default: parquet
        df.to_parquet(save_path, engine="pyarrow")

    return df
```
