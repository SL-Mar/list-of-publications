### 1. Importations

```python
import requests
import json
from IPython.display import display, Markdown

# Replace with your actual webhook URL
webhook_url = "http://localhost:5678/webhook/af822f6e-a7d6-4cb1-bc68-53c967428756"
```

### 2. Model Sonar Reasoning 

```python
response = requests.get(webhook_url, json={})  # Add payload if required, else remove json={}

# Try to parse the response as JSON and display as Markdown
try:
    data = response.json()  # Expecting {"content": "..."}
    markdown_text = data.get('content', '')
    display(Markdown(markdown_text))
except Exception as e:
    print(f"Error parsing response as JSON: {e}")
    print("Raw response:")
    print(response.text)
```
